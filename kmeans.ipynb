```python
from sklearn.cluster import k_means
from matplotlib import pyplot as plt
import json
import numpy as np
```

```python
img = plt.imread('img/3a6d449c3fd40be1a94d3e94b303f365.jpg')/255
plt.imshow(img)
```

```python
centers, y_hat, _ = k_means(img.reshape(-1, 3), 10)

img_compressed = centers[y_hat].reshape(img.shape)

plt.imshow(img_compressed)
```

```python
colors = centers[np.argsort(np.sum(centers, axis=1))]
plt.scatter([0]*len(centers), np.arange(len(centers)), c=colors, s=1000, marker='s')
```

```python
with open('./colors/blues.json', 'w+') as _:
    json.dump(colors.tolist(), _)
```
